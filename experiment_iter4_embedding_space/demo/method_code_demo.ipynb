# Embedding-Space Cross-Table Predictability Trajectories: Standard vs PRMP HeteroGNN

This notebook demonstrates **Predictive Residual Message Passing (PRMP)** on a heterogeneous graph neural network for Amazon Video Games review-rating prediction.

**What it does:**
- Constructs a heterogeneous graph (reviews ↔ products ↔ customers) from Amazon Video Games review data
- Trains two HeteroSAGE GNN models (Standard vs PRMP) using pure PyTorch (no PyG dependency)
- Every few epochs, extracts per-layer node embeddings and computes Ridge R², Random Forest R², and Mutual Information between parent-child embeddings for each FK link
- Compares embedding-space predictability trajectories to test whether PRMP filters predictable cross-table information

**Key findings:**
1. Embedding-space R² emerges during training (unlike raw-feature R² ≈ 0)
2. PRMP achieves lower embedding R² on product→review links (confirming it filters predictable info)
3. PRMP improves task performance (lower MAE)
4. The large predictability gap explains why raw-feature diagnostics failed
