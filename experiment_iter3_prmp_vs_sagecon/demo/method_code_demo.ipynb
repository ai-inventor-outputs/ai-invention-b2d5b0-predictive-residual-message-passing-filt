# PRMP vs SAGEConv Benchmark on RelBench rel-hm Tasks

**Predictive Residual Message Passing (PRMP)** benchmarked against standard HeteroSAGEConv on 2 RelBench rel-hm tasks:
- **user-churn** (classification, AUROC)
- **item-sales** (regression, MAE)

Three model variants compared per-instance on the test set:
- **(A) Standard SAGEConv** — baseline (`predict_baseline`)
- **(B) PRMP with learned predictions** — our method (`predict_our_method`)
- **(C) PRMP with random frozen predictions** — ablation (`predict_ablation_random`)

This notebook loads pre-computed experiment results and analyzes the performance differences across tasks, seeds, and model variants.
