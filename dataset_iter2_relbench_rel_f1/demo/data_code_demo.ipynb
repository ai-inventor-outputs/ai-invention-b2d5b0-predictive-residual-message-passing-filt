# RelBench rel-f1 (Formula 1) Dataset: FK Links & Cardinality Analysis

This notebook demonstrates the **Ergast Formula 1 relational database** analysis pipeline, which:

1. **Enumerates 22 foreign-key (FK) links** across 14 tables (619K+ rows)
2. **Computes cardinality statistics** (mean, median, max, p95, std) per FK link
3. **Generates aligned parent-child feature matrices** for cross-table predictability analysis
4. **Outputs structured metadata** for PRMP (Predictive Residual Message Passing) hypothesis regime analysis

High-cardinality links (e.g., `lap_times→drivers`: mean=3792, `results→status`: mean=188) are ideal candidates for cross-table predictability computation.

**Data source**: [Ergast Motor Racing Data](http://ergast.com/mrd/) (CC-BY-4.0)
