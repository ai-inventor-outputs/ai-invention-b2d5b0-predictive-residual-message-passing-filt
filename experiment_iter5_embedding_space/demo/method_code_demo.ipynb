# Embedding-Space Cross-Table Predictability on rel-stack

**Third Dataset for Revised Regime Theory**

This notebook demonstrates the embedding-space cross-table predictability experiment on the rel-stack (Stack Exchange) dataset. It:

1. **Defines** the Standard HeteroSAGEConv and PRMP (Predictive Residual Message Passing) HeteroSAGEConv architectures
2. **Loads** pre-computed experiment results (training metrics, embedding R² trajectories)
3. **Analyzes** PRMP vs Standard performance on post-votes (regression) and user-engagement (classification) tasks
4. **Visualizes** training curves, R² trajectories, and task performance comparisons

**Key Finding**: PRMP significantly outperforms Standard on both tasks (post-votes MAE: 0.073 vs 0.564; user-engagement AUROC: 0.767 vs 0.505).
