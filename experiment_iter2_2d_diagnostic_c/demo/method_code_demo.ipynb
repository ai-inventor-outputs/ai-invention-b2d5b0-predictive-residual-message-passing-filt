# 2D Diagnostic: Cross-Table Predictability (R²) and Cardinality for All FK Links

This notebook demonstrates a **2D FK diagnostic** that computes cross-table predictability (R², Ridge 5-fold CV) and mutual information for foreign-key links across relational datasets.

**What it does:**
- Loads aligned parent/child feature matrices for FK links across 3 RelBench datasets (rel-hm, rel-stack, rel-amazon)
- Computes **Ridge regression R²** (linear, 5-fold CV) and **Random Forest R²** (nonlinear baseline) for each FK link
- Computes **mutual information** between parent and child features
- Generates 2D diagnostic scatter plots (cardinality vs R², cardinality vs MI, Ridge vs RF)
- Classifies FK links into 4 regimes based on cardinality and predictability thresholds

**Key findings (from full analysis):** Spearman(card, R²)=0.68 (p=0.01), Ridge-RF correlation=0.87, article→transaction shows strong nonlinear signal (R²_rf=0.25 vs R²_ridge=0.07).
